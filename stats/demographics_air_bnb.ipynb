```python

import pandas as pd
import numpy as np
import pingouin
from scipy.stats import chi2_contingency

```

```python
demographics_df = pd.read_csv('../data/clean/demographics.csv', dtype=str, header=[0,1,2,3], index_col=[1,2])
demographics_df
```

```python
import itertools

demographics_df.columns.levels[1]
```

```python
df = pd.melt(demographics_df, ignore_index=False).astype(str)
df.reset_index() 
```

```python
filt = (df.apply(lambda r: r.astype('string').str.contains('Black').any(), axis=1))

black_zip_codes = df[filt]
black_zip_codes = black_zip_codes.sort_index()
black_zip_codes["value"] = black_zip_codes["value"].astype(float)
black_zip_codes.index.names = ["Year", "Zip Code"]
black_zip_codes = black_zip_codes.groupby(level=[0,1]).sum()
black_zip_codes = black_zip_codes.rename(columns={"value": "Black Population"})
black_zip_codes
```

```python
asian_zip_codes_filt = (df.apply(lambda r: r.astype('string').str.contains('Asian').any(), axis=1))
# df[asian_zip_codes_filt].sort_index()

asian_zip_codes = df[asian_zip_codes_filt].sort_index()
asian_zip_codes["value"] = asian_zip_codes["value"].astype(float)
asian_zip_codes.index.names = ["Year", "Zip Code"]
asian_zip_codes = asian_zip_codes.groupby(level=[0,1]).sum()
asian_zip_codes = asian_zip_codes.rename(columns={"value": "Asian Population"})

asian_zip_codes
```

```python
indigenous_zip_codes_filt = (df.apply(lambda r: r.astype('string').str.contains('Native').any(), axis=1))
# df[indigenous_zip_codes_filt].sort_index()

indigenous_zip_codes = df[indigenous_zip_codes_filt].sort_index()
indigenous_zip_codes["value"] = indigenous_zip_codes["value"].astype(float)
indigenous_zip_codes.index.names = ["Year", "Zip Code"]
indigenous_zip_codes = indigenous_zip_codes.groupby(level=[0,1]).sum()
indigenous_zip_codes = indigenous_zip_codes.rename(columns={"value": "Native American Population"})

indigenous_zip_codes
```

```python
# white
white_zip_codes_filt = (df.apply(lambda r: r.astype('string').str.contains('White').any(), axis=1))
# df[white_zip_codes_filt].sort_index()

white_zip_codes = df[white_zip_codes_filt].sort_index()
white_zip_codes["value"] = white_zip_codes["value"].astype(float)
white_zip_codes.index.names = ["Year", "Zip Code"]
white_zip_codes = white_zip_codes.groupby(level=[0,1]).sum()
white_zip_codes = white_zip_codes.rename(columns={"value": "White Population"})

white_zip_codes
```

```python
# latino

latinx_zip_codes = df[df['variable_1'] == 'Hispanic or Latino (of any race)'].sort_index()
latinx_zip_codes["value"] = latinx_zip_codes["value"].astype(float)
latinx_zip_codes.index.names = ["Year", "Zip Code"]
latinx_zip_codes = latinx_zip_codes.groupby(level=[0,1]).sum()
latinx_zip_codes = latinx_zip_codes.rename(columns={"value": "Latinx Population"})

latinx_zip_codes
```

```python
race_demographics_pop_count_df = pd.concat([latinx_zip_codes, white_zip_codes, indigenous_zip_codes, asian_zip_codes, 
           black_zip_codes], axis=1).sort_index()

race_demographics_pop_count_df
```

```python
# get air bnb lists
air_bnb_df = pd.read_csv('../data/clean/inside_airbnb_clean.gz', low_memory=False)
air_bnb_df["borough"] =  air_bnb_df["borough"].replace(' ', '_', regex=True)
air_bnb_df = air_bnb_df.set_index(["year", "zipcode"])
air_bnb_df.sort_index()
```

```python
# todo: check if air bnb is correct

air_bnb_df = air_bnb_df.sort_index().groupby(level=[0,1]).count().rename(columns = {'id': 'Number of Air BnB Listings'})
air_bnb_df = air_bnb_df["Number of Air BnB Listings"].to_frame()
air_bnb_df
```

```python
race_v_airbnb_df = pd.concat([race_demographics_pop_count_df, air_bnb_df], axis=1).sort_index()

race_v_airbnb_df = race_v_airbnb_df.fillna(0)
race_v_airbnb_df
```

```python
race_v_airbnb_df.to_csv('../stats/zipcode_data/mapped_data/demographic_race_v_airbnb.csv')
```
