```python
# import libraries

import pandas as pd
import numpy as np
import pingouin
from scipy.stats import chi2_contingency
```

```python
# import zillow home value data 

df = pd.read_csv('data/clean/zhome_value.csv')
df["year"] = pd.to_datetime(df.year.astype(str)).dt.year
df.head(10)
```

```python
# Split by before and after 2015 

post_airbnb_df = df[df["year"] >= 2015].sort_values(by="year", ascending=True)
pre_airbnb_df = df[df["year"] < 2015].sort_values(by="year", ascending=True)
```

```python
# Test whether there is a statistically significant difference
#     between home values before 2015 and home values after 2015 (inclusively).

p = pingouin.ttest(pre_airbnb_df["home_value"],  post_airbnb_df["home_value"])['p-val'][0]
p = float(p)
sig = p < 0.05

sig
```
