```python
#Loading packages 

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#df = pd.read_csv('data.csv')

zhome_values_zip = pd.read_csv('zhome_value_zip_df.csv')

zhome_values_zip = pd.DataFrame(zhome_values_zip)

#print(zillow_home_values_data)
```

```python
# z stands for zillow

zhome_values_zip
```

```python
zhome_values_zip.columns
```

```python
# df = df.drop(df.columns[[0, 1, 3]], axis=1)

#dropping columns from data frame we dont need 8 = jan 2000. 31 = dec 2002. 0 = Regiontype. 4=statename. 6= metro. 

zhome_values_zip = zhome_values_zip.drop(zhome_values_zip.columns[9:33], axis=1)
zhome_values_zip = zhome_values_zip.drop(zhome_values_zip.columns[[0,1,3,4,7,]], axis=1)
```

```python
#new columns 
zhome_values_zip.columns
```

```python
#new dataframe

zhome_values_zip
```

```python
# creating new dataframe with the name of the columns replaced

#df2 = df.rename({'a': 'X', 'b': 'Y'}, axis=1)

new_column_df = zhome_values_zip.rename({'RegionName' : 'zip', 'State' : 'state',
                                           'City' : 'city', 'CountyName' : 'county_name'}, axis = 1)
```

```python
#new dataframe with columns renamed

new_column_df
```

```python
#df = df.melt(id_vars=["location", "name"], 
#         var_name="Date", 
#         value_name="Value"

#replacing the '2002-01-31 colums to 'Date' and assiging their values.

zip_df = new_column_df.melt(id_vars=['zip', 'state', 'city', 'county_name'], 
                                                       var_name= "date", 
                                                       value_name= "home_value")

```

```python
#new columns with dates as rows

zip_df.columns
```

```python
#changing date column to datetime

zip_df['date'] = pd.to_datetime(zip_df['date'])
```

```python
#    li = ['Albert', 'Louis', 'John']
#    df[df.name.isin(li)]

#creating new df where only the word NY pulls from the State column.

ny_column = ['NY']

zip_df = zip_df[zip_df.state.isin(ny_column)]
```

```python
#new data frame with the column, state, only has NY

zip_df
```

```python
#selecting only 5 boro counties in county_name column

five_boros = zip_df[zip_df['county_name'].isin(['Queens County', 'New York County', 'Bronx County', 'Richmond County', 'Kings County'])]
```
